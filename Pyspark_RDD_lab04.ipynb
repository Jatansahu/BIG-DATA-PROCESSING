#– TASK-1: Generate 100 random numbers in range 0 to 10 using numpy
#randint function with the seed set to 10. Create a RDD using the paral-
#lelize function using data generated in previous step. Calculate the fre-
#quency of each number (0 - 10) using appropriate function of RDD.

```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
import numpy as np
np.random.seed(10)
ran_num  = np.random.randint(10 , size= 100 )
ran_num
```

```python
# Create SparkContext
sc = SparkContext("local", "Frequency of each number(0-10)")
```

```python
# Create RDD from list of random numbers
rdd2 = sc.parallelize(ran_num)
```

```python
# Calculate frequency of each number using countByValue() method
freq = rdd2.countByValue()

#printing frequeancy of each number 
for number , freq in sorted(freq.items()):
  print(f"Number {number} appears {freq} times")
```

#TASK-2: In this task you will calculate the frequency of each word in
#text8 dataset mentioned above. Create a RDD using the text8 dataset.

#Use appropriate functions of the RDD to get the word frequencies. Fil-
#ter the RDD using appropriate function to get the frequencies of words

#containing the letter ’a’.

```python
#For running this first disconnect all and run only task 2
#Import text8 file
```

```python
# # Install pyspark
#!pip install pyspark
```

```python
# Importing libraries
import pandas as pd
import numpy as np
from pyspark import SparkConf, SparkContext
```

```python
# Create a Spark_conf object and seting the application name
spark_conf = SparkConf().setAppName("Frequencies")

# Creating  object
sc = SparkContext.getOrCreate(conf=spark_conf)
# sc = SparkContext(conf=spark_conf)


#Importing text8 data
rddd1 = sc.textFile("/content/drive/MyDrive/BDP/datasets/text8")

#Transformations are lazy operations, instead of updating an RDD, return another RDD


#flatMap is needed here to return every word (separated by a space) in the line as an Array
rddd2 = rddd1.flatMap(lambda line: line.split(" "))
```

```python
#Print rdd2 result to console
#print(rdd2.collect())
```

```python
# Maping each words, and using reduceByKey to get the frequency of each word
#reduceByKey -->  get an RDD of the count of every unique word by aggregating (adding up) all the 1's you wrote in the last step
#map --> map each word to a value of 1 so they can be summed
rdd_freq = rddd2.map(lambda word: (word, 1)).reduceByKey(lambda a, b: a + b)

# Filtering  the RDD to get the frequencies of words containing the letter 'a'
rdd_freq_a = rdd_freq.filter(lambda word_freq: 'a' in word_freq[0])

# Printing the collected result
result = rdd_freq_a.collect()
for i, j in result:
    print(i,j)
```

```python
from pyspark.sql import SparkSession
from pyspark import SparkContext
```

```python
# Implementing JSON File in PySpark

spark = SparkSession.builder \
    .master("local[1]") \
    .appName("PySpark Read JSON") \
    .getOrCreate()

# Reading JSON file into dataframe    
dataframe = spark.read.json("/content/drive/MyDrive/BDP/datasets/iris.json")
dataframe.printSchema()
dataframe.show()

```

```python
# Calculate Pearson Correlation between the columns petal- Length and petalWidth
from pyspark.sql.functions import *
dataframe.corr('petalLength','petalWidth')
```

We can see from the above results that the correlation between petalLength and petalWidth is a positive value which means they are positively related.

Show the columns sepalLength, sepalWidth and species for the rows of data that has petalLength greater than or equal to 1.4 using the appropriate dataframe API.

```python
dataframe.filter(dataframe.petalLength >=1.4).select("sepalLength",'sepalWidth','species').show()
```
